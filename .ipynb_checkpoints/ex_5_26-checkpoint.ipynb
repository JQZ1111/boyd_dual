```python
import jax 
import jax.numpy as jnp
import numpy as np
import matplotlib.pyplot as plt
import jaxopt
```

```python
# pathological case: 
def f(multipliers, k):
    a, b = multipliers
    Q = (1 + a + b)*jnp.eye(2)
    P = jnp.array([-2*a - 2*b, -2*a + 2*b])
    x = -0.5*jnp.linalg.inv(Q)@P
    D = x.T@Q@x + P@x + a + b
    return -(D - k*((1 + a + b)**2)*((1 + a + b) < 0)), x
```

```python
x = f(multipliers, 0)[1]
multipliers
```

```python
optimizer = jaxopt.GradientDescent(f, maxiter = 1)
multipliers = jnp.array([1., 1.])
for i in range(10):
    for j in range(40):
        print(-f(multipliers, 0))
        multipliers, state = optimizer.run(multipliers, (10**i)*0.1)
```
